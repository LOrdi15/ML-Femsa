```python
import pandas as pd
```

```python
data_client = pd.read_csv('Client.csv')
```

```python
data_client
```

```python
data_client = data_client.drop("NAME", axis= 1 )
data_client = data_client.drop('CONTACT_NAME', axis=1)
data_client = data_client.drop('PHONE', axis=1)
data_client = data_client.drop('EMAIL', axis=1)
data_client = data_client.drop("STORE_URL", axis=1)
data_client = data_client.drop("NUTENTA_ID", axis=1)
data_client = data_client.drop("STORE_SIZE", axis=1)
data_client = data_client.drop("ALCOHOL_ALLOWED", axis=1)
data_client = data_client.drop("FRIDGE_NUMBER", axis=1)
data_client = data_client.drop("BURO", axis=1)
data_client = data_client.drop("CONCESSION", axis=1)
data_client = data_client.drop("CONCESSION_TIER", axis=1)
data_client = data_client.drop("REJECTED_REASON", axis=1) 
data_client["APPLICATION_STATUS"] = data_client["APPLICATION_STATUS"].replace({'PENDING': 0, 'APPROVED': 1, 'REJECTED': -1})
data_client["ACTIVE"] = data_client["ACTIVE"].replace({False: 0, True:1})
data_client['VISIT_DAYS'] = data_client['VISIT_DAYS'].fillna(0)
data_client['CLIENT_NUMBER'] = data_client['CLIENT_NUMBER'].fillna(0)
data_client['ROUTE_ID'] = data_client['ROUTE_ID'].fillna(0)
data_client['CREATED_BY'] = data_client['CREATED_BY'].fillna(0)
data_client['UPDATED_BY'] = data_client['UPDATED_BY'].fillna(0)



```

```python
data_client.to_csv("data_client_limpio.csv", index=False)
```
